```python
# Install the required libraries first
!pip install sentence-transformers seaborn matplotlib scikit-learn

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Write 10 sentences across exactly 3 topics
sentences = [
    # Topic 1: Cricket
    "Maintaining a low economy rate is crucial for a spin bowler in T20 matches.",
    "The fast bowler improved his career bowling average after taking a five-wicket haul.",
    "A high strike rate means the batsman scores quickly off the deliveries faced.",
    "He delivered a perfect yorker to completely dismantle the stumps.",

    # Topic 2: Cooking
    "Preheat the oven to 180°C before baking the chocolate cake.",
    "Dice the onions finely and sauté them until they are golden brown.",
    "Add a pinch of salt and simmer the tomato sauce for twenty minutes.",

    # Topic 3: Cybersecurity
    "The firewall successfully blocked the incoming DDoS attack on the server.",
    "Always use two-factor authentication to secure your online banking accounts.",
    "The security analyst detected a vulnerability in the network protocol."
]

# 2. Generate embeddings using the required model
print("Loading model and generating embeddings...")
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)

# 3. Compute and display a 10x10 cosine similarity matrix as a heatmap
similarity_matrix = cosine_similarity(embeddings)

plt.figure(figsize=(10, 8))
sns.heatmap(similarity_matrix, annot=True, cmap="YlGnBu", fmt=".2f",
            xticklabels=[f"S{i+1}" for i in range(10)],
            yticklabels=[f"S{i+1}" for i in range(10)])
plt.title("Cosine Similarity Matrix of 10 Sentences")
plt.show()

# 4. Query sentence similarity
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

# Compute similarity between query and all 10 sentences
query_similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get the indices of the top 2 most similar sentences
top_2_indices = np.argsort(query_similarities)[-2:][::-1]

print(f"\nQuery: '{query}'\n")
print("Top 2 most similar sentences:")
for i, idx in enumerate(top_2_indices):
    print(f"{i+1}. {sentences[idx]} (Score: {query_similarities[idx]:.4f})")
```
